```python
#3.2
from scipy.constants import G
from math import sqrt

M = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
R  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
N  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

S = zip(M, R, N)   

V = []
for d in S:
    m, r, n = d
    v = sqrt(2*G*m/(r*1000)) #(radius from km to m)
    V.append(v)
    print(f"{n}: mass = {m:.2e} kg")

dictionary = dict(zip(N, V))
print()
print(dictionary)
```

```python
#3.10

def cylinder(r, h):
    from math import pi
    A = 2 * pi * (r** + r*h)
    V = pi * h * r**2
    return A, V

A, V = cylinder(5, 15)
ratio = A/V
print(ratio)
```

```python
#3.18
import numpy as np
from IPython.display import display, Markdown

d = [18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
v = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]

d = np.array(d)
v = np.array(v)

b = (np.sum((d - np.mean(d)) * (v - np.mean(v)))) / np.sum((d - np.mean(d))**2)
a = np.mean(v) -  b * np.mean(d)

display(Markdown(f"$v = {a:.2f} + {b:.2f} \\, d$"))
```

```python
#3.22
import matplotlib.pyplot as plt
from math import e
import numpy as np

print(f"using Python lists:")
X = list(range(-5, 6))
Y = []

for x in X:
    y = ( e**x + e**(-x) ) / 2
    Y.append(y)

plt.subplots()
plt.scatter(X, Y)
plt.xlabel('x')
plt.ylabel('cosh(x)')
plt.grid(True)
plt.title('catenary')
plt.show()

print(f"using Numpy:")

x = np.linspace(-5, 5, num = 1000)
y = ( e**x + e**(-x) ) / 2
plt.subplots()
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('cosh(x)')
plt.grid(True)
plt.title('catenary')
plt.show()
```
