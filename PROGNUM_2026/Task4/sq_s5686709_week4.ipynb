```python
#4.12

from math import sin, cos, radians
import numpy as np
from matplotlib import pyplot as plt

t = radians(30)
R_t = cos(t) + 1j*sin(t)

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

z = x + 1j*y
z_t = z * R_t
xt = np.array(z_t.real)
yt = np.array(z_t.imag)

x2 = x - 2
y2 = y - 2
z2 = x2 + 1j*y2
zt = z2 * R_t
x_t = np.array(zt.real)
y_t = np.array(zt.imag)

plt.plot(x, y)
plt.plot(xt, yt, color = 'red')
plt.plot(x_t, y_t, color = 'green')
plt.gca().set_aspect(1)
plt.show()
```

```python
#4.14

def funcfilter(func, seq):
    import numpy as np
    X = np.asarray(seq)
    x = X[X > 0]  #filtered
    return func(x)

print(funcfilter(np.log10, np.array([-100, -10, 0, 10, 100])))

print(funcfilter(np.log10, 100))
```

```python
#4.16
from scipy.integrate import quad
from math import sqrt
from astropy.cosmology import Planck18

z = 2 

H0 = Planck18.H0.value #km/(Mpc s)
Om0 = Planck18.Om0 #no units
Ode0 = Planck18.Ode0 #no units
print(Ode0)

#find tl(z)
#f = 1/((1+x) * H0 * sqrt(Om0 * (1+x)**3 + Ode0))
def zfunc(x):
    return 1/((1+x) * (H0 * 3.1536e16 / 3.0857e19) * sqrt(Om0 * (1+x)**3 + Ode0)) 
#above, H0 multiplied like that to get the units to Gyr
i, err = quad(zfunc, 0, z)

print(f"z = {i} Gyr")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

x = np.random.normal(loc=0.0, scale=1, size=100000)
y = np.random.normal(loc=2.0, scale=1, size=100000)

H, xedges, yedges, im = plt.hist2d(x, y, bins=30, cmap='Blues')  
plt.colorbar(im)
#notes:xedges and yedges r the arrays of the edges of the bins along the respective axes
#H is a 2D array of shape (ny, nx) containing counts/density in each bin

#for contour: grid of points at the center of each bin;
# Compute bin centers:
X, Y = np.meshgrid((xedges[1:] + xedges[:-1])/2, #(averages)
                   (yedges[1:] + yedges[:-1])/2)
#basically meshgrid(centers) makes a grid
#X, Y = np.meshgrid(x, y)

# Flatten histogram and sort
Hflat = H.flatten()
Hsort = np.sort(Hflat)[::-1]

# cumulative distribution
cdf = np.cumsum(Hsort)
cdf = cdf / cdf[-1]

# find density levels for 1σ, 2σ, 3σ
level1 = Hsort[np.searchsorted(cdf, 0.68)]
level2 = Hsort[np.searchsorted(cdf, 0.95)]
level3 = Hsort[np.searchsorted(cdf, 0.997)]

levels = [level3, level2, level1]

plt.contour(X, Y, H.T, levels=levels, colors='red', linewidths=1.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.title("2D Histogram + contour")


plt.show()
```
